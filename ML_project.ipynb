```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
new_df['fueltype_le']=le.fit_transform(new_df[['fueltype']])
new_df['aspiration_le']=le.fit_transform(new_df[['aspiration']])
new_df['doornumber_le']=le.fit_transform(new_df[['doornumber']])
new_df['carbody_le']=le.fit_transform(new_df[['carbody']])
new_df['drivewheel_le']=le.fit_transform(new_df[['drivewheel']])
new_df['enginelocation_le']=le.fit_transform(new_df[['enginelocation']])
new_df['enginetype_le']=le.fit_transform(new_df[['enginetype']])
new_df['cylindernumber_le']=le.fit_transform(new_df[['cylindernumber']])
new_df['fuelsystem_le']=le.fit_transform(new_df[['fuelsystem']])
```

```python
dfs=new_df.select_dtypes(np.number)
dfs.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, SGDRegressor, Lasso, Ridge, ElasticNet
from sklearn.metrics import r2_score, mean_squared_error
import statsmodels.api as sma
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SequentialFeatureSelector
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.stats.api as ssa
from statsmodels.stats.stattools import durbin_watson
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
X=dfs.drop('price',axis=1)
y=dfs['price']
xtrain,xtest,ytrain,ytest=train_test_split(X,y,random_state=10,test_size=0.3)
```

```python
models={'Linear regression':LinearRegression(),'SGD regression':SGDRegressor(),'Random forest':RandomForestRegressor(),'Decision tree':DecisionTreeRegressor(),'K Neighbors':KNeighborsRegressor()}
report=pd.DataFrame(columns=['Model name','Train error','Test error','Rsquare train','Rsquare test'])
for model,model_name in list(zip(models.values(),models.keys())):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtrain)
    ypred_t=model.predict(xtest)
    print('Fitting the model...',model_name)
    val=pd.DataFrame({'Model name':model_name,'Train error':f'{mean_squared_error(ytrain,ypred,squared=False)}','Test error':f'{mean_squared_error(ytest,ypred_t,squared=False)}','Rsquare train':f'{r2_score(ytrain,ypred)}','Rsquare test':f'{r2_score(ytest,ypred_t)}'},index=[0])
    report=pd.concat([val,report],ignore_index=True)

report
```
